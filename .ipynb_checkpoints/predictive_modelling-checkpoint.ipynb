# Feature engineering

```python
import pandas as pd
import numpy as np
import gc
import xgboost as xg
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
```

```python

movies_info=pd.read_csv("./datasets/movie_data_merged_v2.csv").drop("Unnamed: 0",axis=1)
user_info=pd.read_csv(r'./datasets/users.csv')
movie_ratings = pd.read_csv(r'./datasets/user_ratings.csv')
```

```python
ratings_v1=movie_ratings.merge(user_info,how="left",left_on="user",right_on="userid")
ratings_v2=ratings_v1.merge(movies_info,how="inner",left_on="movie",right_on="movie_id")
print(ratings_v2.shape)
print(movie_ratings.shape)
###loss of less than 5% of all ratings
```

### bucketize ratings into 1-3 & 4-5 as 0 and 1

```python
ratings_v2["reccomendation_bucket"]=ratings_v2["rating"].apply(lambda x: 0 if x<=3 else 1)
```

## train validation split. 8:1:1

```python
#shuffle rows
ratings_v2 = ratings_v2.sample(frac=1, axis=0,random_state=1).reset_index(drop=True)
n_rows=ratings_v2.shape[0]
one_tenth=n_rows//10
##train
train = ratings_v2.iloc[:8*one_tenth,]
validation=ratings_v2.iloc[8*one_tenth:9*one_tenth,]
test=ratings_v2.iloc[9*one_tenth:,]
```

## OHE of genre

```python
### split genre into ohe
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

genres=train["genre_label"].apply(lambda x: x.replace("|"," | "))
genre_split=CountVectorizer(token_pattern=r"[a-z'-]+")
genre_ohe_sparse=genre_split.fit_transform(train["genre_label"])#sparse

genre_labels=list(map(lambda x: "genre_"+x,genre_split.get_feature_names()))

#genre_ohe=genre_ohe_sparse.todense()#dense
#genre_ohe=pd.DataFrame(genre_ohe,columns=genre_labels)
#not yet concatenated to ratings to make things tidier
```

## OHE of user occupation, age bracket and gender

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
age_encoder = OneHotEncoder()
age_ohe=age_encoder.fit_transform(train[['age']])##still in sparse form
age_labels=list(age_encoder.active_features_)
age_labels=list(map(lambda x:"age_"+str(x),age_labels))
```

```python
occupation_encoder= OneHotEncoder()
occupation_ohe=occupation_encoder.fit_transform(train[['occupation']])##still in sparse form
occupation_labels=list(occupation_encoder.active_features_)


occupation_mapping={0:"other/unspecified",1:"academic/educator",2:"artist", 3:"clerical/admin",4:"college/grad student",
            5:"customer service", 6:"doctor/health care", 7:"executive/managerial",8:"farmer",9:"homemaker",
            10:"K-12 student", 11:"lawyer",12:"programmer",13:"retired",14:"sales/marketing",15:"scientist",
            16:  "self-employed",17:"technician/engineer",18:"tradesman/craftsman",19:"unemployed",20:"writer"}

occupation_labels=list(map(lambda x:"occupation_"+occupation_mapping[x],occupation_labels))
```

```python
gender_labeler=LabelEncoder()
gender_ohe=gender_labeler.fit_transform(train["gender"])
train["is_male"]=gender_ohe
```

## Zip-code

```python
zip_encoder= OneHotEncoder()
train["zip_cat"]=train["zip_code"].apply(lambda x: x[0])
zip_type_OHE=zip_encoder.fit_transform(train[["zip_cat"]])
```

## country

```python
# entities to merge/seperate.
# uk, britain, british, united kingdown
#american, united states
#Germany, west germany, weimar republic
#india, bollywood, bengali
#russia, soviet union
#czechslovakia, czech republic
# yugoslavia? -> serbia/
import re

def country_mapping(country):
    country=re.sub(r"(UK)|(British)|(Great Britain)","United Kingdom",country)
    country=re.sub(r"(USA?)|(American)|(America)","United States",country)
    country= re.sub(r"(Filming locations)|(The)|(FR )","",country)
    country=re.sub(r"(Australian)","Australia",country)
    country=re.sub(r"(Canadian)","Canada",country)
    country=re.sub(r"(Chinese)","China",country)
    country=re.sub(r"(Yugoslavia)","Serbia",country)
    country=re.sub(r"(Czechoslovakia)","Czech Republic, Slovakia",country)
    country=re.sub(r"(Weimar Republic( of Germany)?)|(West Germany)","Germany",country)
    country=re.sub(r"(Bollywood)|(Bengali)","India",country)
    country=re.sub(r"(Japanese)","Japan",country)
    country=re.sub(r"(Soviet Union)","Russian",country)
    return country

#token pattern r"([a-zA-Z]{3,} [a-zA-Z]+)|([a-zA-Z_]+)"
#ratings_v2["Origin/Ethnicity"].unique()
```

```python
country_encoder=CountVectorizer(token_pattern=r"([a-zA-Z]{3,} [a-zA-Z]+)|([a-zA-Z_]+)"
                                ,min_df=1
                               )
country_feat_train=country_encoder.fit_transform(train["Origin/Ethnicity"].fillna("").apply(country_mapping))
```

## Movie tags

```python
#extract tags that has appeared in at least 1% or 30 movies
tag_vectorizers=CountVectorizer(token_pattern=r"[a-z- &']{3,}"
                                ,min_df=0.01)
tag_vectorizers.fit(train["tags_clean"].unique())
tag_sparse=tag_vectorizers.transform(train["tags_clean"])
tags_labels = list(map(lambda x:"tags_"+x,tag_vectorizers.get_feature_names()))
```

## actors and directors

```python
#actors that have appeared in at least 5 movies
cast_vectorizer=CountVectorizer(min_df=5)
cast_vectorizer.fit(movies_info["Cast_clean"])

#directors that directed at least 3 movies
director_vectorizer=CountVectorizer(min_df=3)
director_vectorizer.fit(movies_info["Director_clean"])


directors_sparse = director_vectorizer.transform(train["Director_clean"])
cast_sparse = cast_vectorizer.transform(train["Cast_clean"])
```

```python
print(len(cast_vectorizer.get_feature_names()))
print(len(director_vectorizer.get_feature_names()))
#train["Director_clean"]
```

## tfidf

```python
# download stopwords if needed
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# using nltk for lemmatization instead
import nltk
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
  
  
nltk_stopwords=set(stopwords.words('english'))
nltk_stopwords.update(["/n","/r","/r/n"])
lmtzr = WordNetLemmatizer()

def nltk_lemmatize(plot):
    cleaned=[]
    for word in plot.split(" "):
        if word in nltk_stopwords:
            continue
        cleaned.append(lmtzr.lemmatize(word))
    return " ".join(cleaned)
```

```python
##clean each unique plot before rejoining back
unique_plot_train=train.groupby(["Plot","movie"],as_index=False).count()[["Plot","movie"]]
unique_plot_train["clean_plot"]=unique_plot_train["Plot"].apply(nltk_lemmatize)
train=pd.merge(left=train, right=unique_plot_train,on="movie", how="left")
```

```python
# fit idf weights to each unique movie plot, instead of by rating.
# sublinear_tf will be used to account for differing plot write-up lengths
plot_vectorizer = TfidfVectorizer(sublinear_tf=True, max_df=0.4,
                                  min_df=0.005)
plot_vectorizer.fit(pd.unique(train["clean_plot"]))
```

```python
plot_vectors = plot_vectorizer.transform(train["clean_plot"])
```

## combine features for modeling

```python
import scipy.sparse as sp

dense_features = train[["is_male","year"]]
X_train=sp.hstack([dense_features,zip_type_OHE,age_ohe,occupation_ohe
                   ,country_feat_train,genre_ohe_sparse,tag_sparse,cast_sparse,directors_sparse
                   ,plot_vectors
                  ])
y_train=train["reccomendation_bucket"]
#y_train=train["rating"]
```

```python
print(dense_features.shape)
print(age_ohe.shape)
print(occupation_ohe.shape)
print(country_feat_train.shape)
print(genre_ohe_sparse.shape)
print(cast_sparse.shape)
print(directors_sparse.shape)
print(X_train.shape)
```

## Data Pipe

```python
## replicate data transformation steps for validation and test.
## assume dataset has same columns as training set originally
## data transformers should have been fitted for training already
def data_pipe(dataset):
    dataset["reccomendation_bucket"]=dataset["rating"].apply(lambda x: 0 if x<=3 else 1)
    genres=dataset["genre_label"].apply(lambda x: x.replace("|"," | "))
    genre_feat=genre_split.transform(genres)
    age_feat=age_encoder.transform(dataset[['age']])
    occupation_feat=occupation_encoder.transform(dataset[['occupation']])
    dataset["is_male"]=gender_labeler.transform(dataset["gender"])
    country_feat=country_encoder.transform(dataset["Origin/Ethnicity"].fillna("").apply(country_mapping))

    unique_plot=dataset.groupby(["Plot","movie"],as_index=False).count()[["Plot","movie"]]
    unique_plot["clean_plot"]=unique_plot["Plot"].apply(nltk_lemmatize)
    dataset=pd.merge(left=dataset, right=unique_plot[["movie","clean_plot"]],on="movie", how="left")
    plot_feat= plot_vectorizer.transform(dataset["clean_plot"])
    
    dataset["zip_cat"]=dataset["zip_code"].apply(lambda x: x[0])
    zip_type_ohe=zip_encoder.fit_transform(dataset[["zip_cat"]])
    tags = tag_vectorizers.transform(dataset["tags_clean"])
    directors= director_vectorizer.transform(dataset["Director_clean"])
    cast=cast_vectorizer.transform(dataset["Cast_clean"])
    x_feat=sp.hstack([dataset[["is_male","year"]],zip_type_ohe,
                      age_feat,occupation_feat,country_feat,genre_feat,tags,cast,directors
                      ,plot_feat
                     ])
    y=dataset["reccomendation_bucket"]
#    y=dataset["rating"]
    return (x_feat,y)    
```

```python
X_test,y_test = data_pipe(test)
X_val,y_val = data_pipe(validation)
```

```python
# arrange feature labels
zip_labels =list(map(lambda x: "zip_cat_"+str(x),list(zip_encoder.active_features_.tolist())))
country_labels=list(map(lambda x: "origin_"+x, 
                        map(lambda x :x[1] if x[0]=="" else x[0], 
                            country_encoder.get_feature_names())))

cast_labels=list(map(lambda x: "cast_"+x,cast_vectorizer.get_feature_names()))

director_labels=list(map(lambda x: "director_"+x,director_vectorizer.get_feature_names()))
plot_vocab=list(map(lambda x: "vocab_"+x,plot_vectorizer.get_feature_names()))

feature_labels=(["is_male","year"]+zip_labels+age_labels+occupation_labels+
                country_labels+genre_labels+
                tags_labels+cast_labels+director_labels+plot_vocab)
```

## naive bayes

## Logistic regression

```python
import sklearn.linear_model

logistic_regression = sklearn.linear_model.LogisticRegression()
logistic_regression.fit(X_train,y_train)

val_pred = logistic_regression.predict(X_val)
val_f1 = sklearn.metrics.f1_score(val_pred,y_val)
val_acc= sklearn.metrics.accuracy_score(val_pred,y_val)
conf_matrix = sklearn.metrics.confusion_matrix(val_pred,y_val)
print(conf_matrix,"\n",val_acc,"\n",val_f1)
```

Naive bayes not that great, with imbalanced predictions

Logistic regression is better. Slightly better accuracy and more balanced predictions.

```python
diagnose=pd.DataFrame([y_val,val_pred]).transpose()
diagnose.columns=["validation","predicted"]
diagnose["FP"]=diagnose.apply(lambda x:x[0]<x[1],axis=1)
diagnose["FN"]=diagnose.apply(lambda x:x[0]>x[1],axis=1)
diagnose["wrong"]=diagnose.apply(lambda x:x[0]!=x[1],axis=1)
```

```python
a=validation.reset_index(drop=True)
wrong_classifications=a[diagnose["wrong"]].join(diagnose[diagnose["wrong"]][["FP","FN"]])
```

```python
import seaborn as sns
fig, axs = plt.subplots(ncols=2,)
fig.set_figwidth(15)
sns.distplot(wrong_classifications["rating"],ax=axs[0]).set_title("misclassified")
sns.distplot(validation["rating"], ax=axs[1]).set_title("original")
```

By nature, all ratings that are 3 are false positives, all ratings that are 4 are false negatives.
Distribution is quite different from original

Likely the model is having problems distinguishing between 3 and 4 ratings

# Gridsearch logs

logs of gridsearch, placed here cause it's too long

```python
tree_depth = [10,15,20,25,30]
min_child_weight=[3,5,7,10]
gamma=[2,3,4,5]
results=[]
for depth in tree_depth:
  for hessians in min_child_weight:
    for g in gamma:
      print((depth,hessians,g))
      tuning=xg.XGBClassifier(max_depth=depth,min_child_weight=hessians,gamma=g, reg_lambda=2,
                             learning_rate=0.3,n_jobs=-1)
      tuning.fit(X_train,y_train,eval_set=[(X_train,y_train),(X_val,y_val)],
              eval_metric='error',verbose=False,early_stopping_rounds=30)
      best_tree=tuning.best_ntree_limit
      best_val_acc=tuning.evals_result()['validation_1']["error"][best_tree-1]
      a="depth: {}, min_child_weight:{}, gamma:{}, best_trees:{}, error:{}".format(depth,
                                                                                     hessians,
                                                                                        g,
                                                                                        best_tree,
                                                                                        best_val_acc
                                                                                        )
      results.append(a)
      print(a)
      #clear memory explicitly 
      del tuning
      gc.collect()
      
```

```python
results
# best perfomers: 
# 'depth: 15, min_child_weight:10, gamma:4, best_trees:50, error:0.301776
# 'depth: 15, min_child_weight:10, gamma:2, best_trees:35, error:0.302919',
# 'depth: 15, min_child_weight:7, gamma:3, best_trees:46, error:0.303085',
# around 15 depth with relatively strong overfitting control tend to perform better
```
